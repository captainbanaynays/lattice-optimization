# Quadratic fitting a plot of lattice enegry v. constant

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from numpy.polynomial.polynomial import Polynomial

run=pd.read_csv('output.csv')
print(run.to_string())
```

```python
plt.scatter(run['aV'].tolist(), run['nrg'].tolist())
x = np.linspace(6.48, 6.78, 256, endpoint = True)

# Use .convert() here because fit shifts the basis of the returned polynomial
regr = Polynomial.fit(run['aV'].tolist(), run['nrg'].tolist(), 2).convert()

print(regr(6.496))
plt.plot(x, regr(x))
plt.xlabel("Lattice Vector Length (Å)")
plt.ylabel("Lattice Energy (eV)")
plt.title("CdTe mp-406 lattice vector length optimization")
plt.show()
```

```python
print(regr)
coefs = regr.coef
print("Lattice constant: ", -coefs[1] / (2 * coefs[2]))
```
